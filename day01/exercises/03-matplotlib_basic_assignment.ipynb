## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python
# line plot using matplotlib
# import matplotlib
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv(r'C:\Users\samir\Desktop\M111-Advanced-Python-and-Database-Management\day01\data\company_sales_data.csv')
data.head()
```

```python
plt.plot(data['month_number'], data['total_profit'])

plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.title('Profit data of last year')
plt.grid(True)  
plt.show()
```

## Exercise 2
**Get total profit of all months and show line plot with the following Style properties**
- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.

## Exercise 9
**Read Bathing soap facewash of all months and display it using the Subplot**

**Expected output**

<img src="../data/Sales_data_of_a_facewash.png" alt="Sales_data_of_a_facewash" style="width: 700px;"/>


```python
_,(ax1,ax2) = plt.subplots(ncols=1,nrows=2)
ax1.plot(data.month_number,data.bathingsoap, color="black",marker=".")
ax1.set_title("sales data of bathing soap")
ax2.plot(data.month_number, data.facewash,color="red",marker=".")
ax2.set_title("sales data of facewash")
plt.xlabel("Month number")
plt.ylabel("Sales units")
plt.show()
```

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

<img src="../data/product_sales_stack_plot.png" alt="product_sales_stack_plot" style="width: 700px;"/>


```python
# write your code here
fig = plt.figure(figsize =(7, 7))

labels = ['Face cream', 'Fase wash', 'Tooth paste','Bathing soap', 'Shampoo', 'Moisturizer']
plt.stackplot(data.month_number, data.facecream, data.facewash, data.toothpaste, data.bathingsoap, data.shampoo, data.moisturizer, labels=labels)
plt.legend(loc = 'upper left')
plt.title('Al product sales data using stack plot')
plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
```
